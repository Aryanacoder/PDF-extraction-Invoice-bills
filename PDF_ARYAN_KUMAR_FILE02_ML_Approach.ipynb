```python
# Import necessary libraries
import csv
import os
import logging
from difflib import SequenceMatcher
import fitz  # PyMuPDF
import cv2
import numpy as np
from pyzbar.pyzbar import decode
import spacy  # Import SpaCy
import re

# Load the SpaCy model
nlp = spacy.load("en_core_web_sm")

# Set up logging
logging.basicConfig(filename='invoice_extraction.log', level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')

# Custom exception for extraction errors
class ExtractionError(Exception):
    """Custom exception for extraction errors"""
    pass

# Function to calculate similarity ratio between two strings
def similar(a, b):
    """Calculate the similarity ratio between two strings."""
    try:
        return SequenceMatcher(None, a, b).ratio()
    except Exception as e:
        logging.error(f"Error in similarity calculation: {str(e)}")
        return 0.0

# Function to extract text and images from a PDF file
def extract_data_from_pdf(pdf_path):
    """Extract text and images from a PDF file."""
    try:
        doc = fitz.open(pdf_path)
        extracted_data = {
            'raw_text': '',
            'file_name': os.path.basename(pdf_path),
            'qr_codes': [],
            'image_texts': []
        }

        for page in doc:
            # Extract text
            extracted_data['raw_text'] += page.get_text()

            # Extract images and QR codes
            image_list = page.get_images(full=True)
            for img_index, img in enumerate(image_list):
                xref = img[0]
                base_image = doc.extract_image(xref)
                image_bytes = base_image["image"]

                # Convert image bytes to numpy array
                nparr = np.frombuffer(image_bytes, np.uint8)
                image = cv2.imdecode(nparr, cv2.IMREAD_COLOR)

                # Check for QR code
                qr_codes = decode(image)
                if qr_codes:
                    for qr_code in qr_codes:
                        extracted_data['qr_codes'].append(qr_code.data.decode('utf-8'))
                else:
                    # If not a QR code, treat as a regular image
                    extracted_data['image_texts'].append(image)

        return extracted_data
    except Exception as e:
        logging.error(f"Error reading PDF {pdf_path}: {str(e)}")
        return {
            'raw_text': '',
            'file_name': os.path.basename(pdf_path),
            'qr_codes': [],
            'image_texts': []
        }

# Function to extract relevant fields using SpaCy NER
def extract_with_ner(text):
    """Extract relevant fields using SpaCy NER with improved accuracy."""
    doc = nlp(text)
    extracted_data = {
        'Invoice Number': '',
        'Invoice Date': '',
        'Due Date': '',
        'Customer Name': '',
        'Phone': '',
        'Total Amount': '',
        'Taxable Amount': '',
        'Round Off': '',
        'Total Discount': '',
        'Total Items / Qty': '',
        'Total amount in words': '',
        'Place of Supply': '',
        'Bank': '',
        'Account #': '',
        'IFSC Code': '',
        'Branch': ''
    }

    # Extract entities using NER
    for ent in doc.ents:
        if ent.label_ == "DATE":
            if not extracted_data['Invoice Date']:
                extracted_data['Invoice Date'] = ent.text
            elif not extracted_data['Due Date']:
                extracted_data['Due Date'] = ent.text
        elif ent.label_ == "MONEY":
            if not extracted_data['Total Amount']:
                extracted_data['Total Amount'] = ent.text
            elif not extracted_data['Taxable Amount']:
                extracted_data['Taxable Amount'] = ent.text
        elif ent.label_ == "ORG":  # Assuming customer name might be recognized as an organization
            extracted_data['Customer Name'] = ent.text
        elif ent.label_ == "PHONE":  # Custom label for phone numbers
            extracted_data['Phone'] = ent.text

    #ner regex pattern
    patterns = {
        'Invoice Number': r"Invoice #: (INV-\d+)",
        'Invoice Date': r"Invoice Date: (\d{2} \w{3} \d{4})",
        'Due Date': r"Due Date: (\d{2} \w{3} \d{4})",
        'Customer Name': r"Customer Details:\n(.*)",
        'Phone': r"Ph:\s*([\d\s+]+)",
        'Place of Supply': r"Place of Supply:\s*(\d+-[A-Z\s]+)",
        'Total Amount': r"Total\s*₹?([\d,]+\.\d{2})",
        'Taxable Amount': r"Taxable Amount\s*₹?([\d,]+\.\d{2})",
        'Round Off': r"Round Off\s*([-]?[\d,]+\.\d{2})",
        'Total Discount': r"Total Discount\s*₹?([\d,]+\.\d{2})",
        'Total Items / Qty': r"Total Items / Qty : (\d+ / [\d.]+)",
        'Total amount in words': r"Total amount \(in words\):\s*(.*)",
        'Bank': r"Bank:\s*(.*)",
        'Account #': r"Account #:\s*(\d+)",
        'IFSC Code': r"IFSC Code:\s*(\w+)",
        'Branch': r"Branch:\s*(.*)"
    }

    # Extract using regex and validate
    for key, pattern in patterns.items():
        match = re.search(pattern, text, re.IGNORECASE | re.MULTILINE)
        if match:
            extracted_data[key] = match.group(1).strip() if len(match.groups()) > 0 else match.group(0).strip()

    # Validate extracted data and calculate trusted accuracy
    trusted_data = {}
    for key, value in extracted_data.items():
        try:
            if value:
                # Define expected format based on the key
                if key == 'Invoice Number':
                    expected_format = f"Invoice Number: {value}"  # Example expected format
                elif key in ['Invoice Date', 'Due Date']:
                    expected_format = f"Date: {value}"  # Example expected format for dates
                elif key == 'Customer Name':
                    expected_format = f"Customer Name: {value}"  # Example expected format for names
                else:
                    expected_format = f"{key}: {value}"  # General expected format

                # Calculate similarity score
                accuracy = similar(expected_format, value)

                # Set dynamic threshold based on the key
                if key in ['Invoice Number', 'Total Amount']:
                    trusted = accuracy > 0.75  # Higher threshold for critical fields
                else:
                    trusted = accuracy > 0.6  # Lower threshold for less critical fields

                # Log low accuracy cases for further analysis
                if not trusted:
                    logging.warning(f"Low accuracy for {key}: {value} (Accuracy: {accuracy:.2f})")

                trusted_data[key] = {
                    'value': value,
                    'accuracy': accuracy,
                    'trusted': trusted
                }
            else:
                trusted_data[key] = {
                    'value': '',
                    'accuracy': 0.0,
                    'trusted': False
                }
        except Exception as e:
            logging.error(f"Error processing {key}: {str(e)}")
            trusted_data[key] = {
                'value': '',
                'accuracy': 0.0,
                'trusted': False
            }

    return trusted_data

# Function to extract item details from the invoice text
def extract_items(text):
    """Extract item details from the invoice text."""
    try:
        # Adjusted item pattern to ensure correct extraction
        item_pattern = r"(\d+)\n(.*?)\n([\d.]+)\n([\d.]+\s*\(.*?\))\n(.*?)\n([\d.]+)\n([\d.]+\s*\(.*?\))\n([\d,.]+)"
        items = re.findall(item_pattern, text, re.MULTILINE)
        if not items:
            logging.warning("No items found in the invoice")
        return items
    except Exception as e:
        logging.error(f"Error extracting items: {str(e)}")
        raise ExtractionError("Failed to extract items")

# Function to save extracted data to a CSV file
def save_to_csv(data, csv_file):
    """Save extracted data to a CSV file."""
    try:
        text = data['raw_text']
        extracted_fields = extract_with_ner(text)

        csv_data = {
            'File Name': data['file_name'],
            'Invoice Number': extracted_fields['Invoice Number']['value'],
            'Invoice Date': extracted_fields['Invoice Date']['value'],
            'Due Date': extracted_fields['Due Date']['value'],
            'Customer Name': extracted_fields['Customer Name']['value'],
            'Phone': extracted_fields['Phone']['value'],
            'Total Amount': extracted_fields['Total Amount']['value'],
            'Taxable Amount': extracted_fields['Taxable Amount']['value'],
            'Round Off': extracted_fields['Round Off']['value'],
            'Total Discount': extracted_fields['Total Discount']['value'],
            'Total Items / Qty': extracted_fields['Total Items / Qty']['value'],
            'Total amount in words': extracted_fields['Total amount in words']['value'],
            'Place of Supply': extracted_fields['Place of Supply']['value'],
            'Bank': extracted_fields['Bank']['value'],
            'Account #': extracted_fields['Account #']['value'],
            'IFSC Code': extracted_fields['IFSC Code']['value'],
            'Branch': extracted_fields['Branch']['value'],
            'QR Codes': '; '.join(data['qr_codes']),
            'Number of Images': len(data['image_texts'])
        }

        # Calculate accuracy and trustworthiness
        for key, value in extracted_fields.items():
            csv_data[f'{key} Accuracy'] = f"{value['accuracy']:.2f}"
            csv_data[f'{key} Trusted'] = str(value['trusted'])

        # Extract item details
        try:
            items = extract_items(text)
            total_items = len(items)  # Count the number of items
            for i in range(1, 6):  # Assuming a maximum of 5 items
                if i <= total_items:
                    item = items[i - 1]
                    csv_data[f'Item {i} Name'] = item[1].strip()
                    csv_data[f'Item {i} Rate'] = item[2]
                    csv_data[f'Item {i} Discounted Rate'] = item[3]
                    csv_data[f'Item {i} Quantity'] = item[4]
                    csv_data[f'Item {i} Taxable Value'] = item[5]
                    csv_data[f'Item {i} Tax Amount'] = item[6]
                    csv_data[f'Item {i} Total Amount'] = item[7]
                else:
                    # If there are fewer items, set the values to 0
                    csv_data[f'Item {i} Name'] = '0'
                    csv_data[f'Item {i} Rate'] = '0'
                    csv_data[f'Item {i} Discounted Rate'] = '0'
                    csv_data[f'Item {i} Quantity'] = '0'
                    csv_data[f'Item {i} Taxable Value'] = '0'
                    csv_data[f'Item {i} Tax Amount'] = '0'
                    csv_data[f'Item {i} Total Amount'] = '0'

            # Add total items count to CSV data
            csv_data['Total Items'] = total_items
        except ExtractionError as ee:
            logging.error(f"Error extracting items: {str(ee)}")

        # Write to CSV
        file_exists = os.path.isfile(csv_file)
        with open(csv_file, 'a', newline='', encoding='utf-8') as f:
            writer = csv.DictWriter(f, fieldnames=csv_data.keys())
            if not file_exists:
                writer.writeheader()
            writer.writerow(csv_data)

        return 1.0  # Return a dummy accuracy for now
    except Exception as e:
        logging.error(f"Error processing file {csv_file}: {str(e)}")
        return 0.0

# Function to process a single invoice PDF and save the extracted data to CSV
def process_invoice(pdf_path, csv_file):
    """Process a single invoice PDF and save the extracted data to CSV."""
    try:
        data = extract_data_from_pdf(pdf_path)
        accuracy = save_to_csv(data, csv_file)
        return accuracy
    except Exception as e:
        logging.error(f"Failed to process {pdf_path}: {str(e)}")
        return 0.0

# Function to process multiple invoice PDFs based on user input
def process_multiple_invoices():
    """Process multiple invoice PDFs based on user input."""
    pdf_files = []
    print("Enter the paths of the PDF files (one per line). Press Enter twice to finish:")
    while True:
        pdf_path = input().strip()
        if pdf_path:
            if os.path.exists(pdf_path) and pdf_path.lower().endswith('.pdf'):
                pdf_files.append(pdf_path)
            else:
                print(f"Invalid path or not a PDF file: {pdf_path}")
        else:
            break

    if not pdf_files:
        print("No valid files entered. Exiting.")
        return

    csv_file = 'extracted_multimulti_19911.csv'
    total_accuracy = 0.0
    for pdf_path in pdf_files:
        total_accuracy += process_invoice(pdf_path, csv_file)

    # Calculate and print the overall accuracy
    if pdf_files:
        overall_accuracy = total_accuracy / len(pdf_files)
        print(f"\nOverall System Accuracy: {overall_accuracy:.2%}")

# Main execution
if __name__ == "__main__":
    process_multiple_invoices()
```
